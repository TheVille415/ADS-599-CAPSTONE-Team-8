```python
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

try:
    from rdkit import Chem
    from rdkit.Chem import Descriptors
    RDKIT_AVAILABLE = True
except Exception:
    RDKIT_AVAILABLE = False
    print("RDKit not available. Chemistry-specific cells may need to be skipped.")
```

## 1. Load raw data

```python
RAW_DATA_PATH = "data/raw/replace_with_file.csv"
df = pd.read_csv(RAW_DATA_PATH)

print("Shape:", df.shape)
df.head()
```

## 2. Cleaning log

```python
cleaning_log = []

def log_step(step, why, rows_before=None, rows_after=None):
    cleaning_log.append({
        "step": step,
        "why": why,
        "rows_before": rows_before,
        "rows_after": rows_after
    })
```

## 3. Standardize column names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
)

log_step(
    step="Standardized column names",
    why="Improves consistency across notebooks and reduces merge problems."
)

df.columns.tolist()
```

## 4. Remove exact duplicates

```python
rows_before = len(df)
df = df.drop_duplicates().copy()
rows_after = len(df)

log_step(
    step="Removed exact duplicate rows",
    why="Duplicate rows can bias summary statistics and downstream modeling.",
    rows_before=rows_before,
    rows_after=rows_after
)

print(f"Removed {rows_before - rows_after} duplicate rows.")
```

## 5. Optional: validate and canonicalize SMILES

```python
SMILES_COL = "canonical_smiles"
```

```python
if RDKIT_AVAILABLE and SMILES_COL in df.columns:
    def smiles_to_mol(x):
        try:
            return Chem.MolFromSmiles(str(x))
        except Exception:
            return None

    rows_before = len(df)
    df["mol"] = df[SMILES_COL].apply(smiles_to_mol)
    df["is_valid_smiles"] = df["mol"].notna()

    invalid_count = (~df["is_valid_smiles"]).sum()
    df = df[df["is_valid_smiles"]].copy()
    rows_after = len(df)

    log_step(
        step="Removed invalid SMILES rows",
        why="Invalid structures cannot support descriptor generation or modeling.",
        rows_before=rows_before,
        rows_after=rows_after
    )

    df["canonical_smiles_clean"] = df["mol"].apply(Chem.MolToSmiles)
    print(f"Removed {invalid_count} invalid SMILES rows.")
else:
    print("Skipping SMILES validation.")
```

```python
if RDKIT_AVAILABLE and "canonical_smiles_clean" in df.columns:
    rows_before = len(df)
    df = df.drop_duplicates(subset=["canonical_smiles_clean"]).copy()
    rows_after = len(df)

    log_step(
        step="Removed duplicate canonical SMILES",
        why="Prevents repeated molecular structures from being overweighted.",
        rows_before=rows_before,
        rows_after=rows_after
    )

    print(f"Removed {rows_before - rows_after} duplicate canonical molecules.")
```

## 6. Missing value handling

```python
missing_before = int(df.isna().sum().sum())

threshold = 0.40
high_missing_cols = df.columns[df.isna().mean() > threshold].tolist()

if high_missing_cols:
    df = df.drop(columns=high_missing_cols)
    log_step(
        step="Dropped high-missingness columns",
        why=f"Columns with more than {int(threshold*100)}% missingness were considered too sparse."
    )

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in numeric_cols:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].median())

missing_after = int(df.isna().sum().sum())
print("Missing values before:", missing_before)
print("Missing values after:", missing_after)
```

## 7. Type fixes

```python
dtype_before = df.dtypes.astype(str)

# Add explicit type conversions below as needed.
# Example:
# df["activity_score"] = pd.to_numeric(df["activity_score"], errors="coerce")

dtype_after = df.dtypes.astype(str)

type_changes = pd.DataFrame({
    "before": dtype_before,
    "after": dtype_after
})

type_changes[type_changes["before"] != type_changes["after"]]
```

## 8. Feature engineering

```python
if RDKIT_AVAILABLE and "mol" in df.columns:
    df["mol_wt"] = df["mol"].apply(Descriptors.MolWt)
    df["logp"] = df["mol"].apply(Descriptors.MolLogP)
    df["h_donors"] = df["mol"].apply(Descriptors.NumHDonors)
    df["h_acceptors"] = df["mol"].apply(Descriptors.NumHAcceptors)
    df["tpsa"] = df["mol"].apply(Descriptors.TPSA)
    df["rotatable_bonds"] = df["mol"].apply(Descriptors.NumRotatableBonds)

    log_step(
        step="Created molecular descriptor features",
        why="Adds structured chemical property features for analysis and modeling."
    )

if "canonical_smiles_clean" in df.columns:
    df["smiles_length"] = df["canonical_smiles_clean"].astype(str).str.len()
    log_step(
        step="Created SMILES length feature",
        why="Captures a basic measure of molecular sequence complexity."
    )

df.head()
```

```python
leakage_checklist = [
    "Exclude identifiers from predictors unless they serve a real analytical purpose.",
    "Do not let label-derived variables enter the model input set.",
    "Fit imputation, scaling, and encoding on training data only.",
    "Check for overlap between training molecules and evaluation molecules."
]

for item in leakage_checklist:
    print("-", item)
```

## 11. Feature selection summary

```python
identifier_cols = [c for c in ["compound_id", "molecule_chembl_id", "pubchem_cid"] if c in df.columns]
candidate_feature_cols = [c for c in ["mol_wt","logp","h_donors","h_acceptors","tpsa","rotatable_bonds","smiles_length"] if c in df.columns]
target_cols = [c for c in ["activity", "label", "target_class"] if c in df.columns]

feature_summary = {
    "identifiers": identifier_cols,
    "candidate_features": candidate_feature_cols,
    "targets": target_cols
}

feature_summary
```
